# Исследование успешности компьютерных игр

**Цели и задачи проекта**<br>
Цель проекта - на основе данных из открытых источников (исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы) выявить закономерности, определяющие успешность игр. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании на следующий год.<br>

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр в США и Канаде. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».Рейтинг игры напечатан на упаковке, содержится в её рекламе и указан на сайте игры.

**Описание данных**
Name — название игры
Platform — платформа
Year_of_Release — год выпуска
Genre — жанр игры
NA_sales — продажи в Северной Америке (миллионы проданных копий)
EU_sales — продажи в Европе (миллионы проданных копий)
JP_sales — продажи в Японии (миллионы проданных копий)
Other_sales — продажи в других странах (миллионы проданных копий)
Critic_Score — оценка критиков (максимум 100)
User_Score — оценка пользователей (максимум 10)
Rating — рейтинг от организации ESRB, которая определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

```python
#приведем названия столбцов к нижнему регистру
data.columns = data.columns.str.lower()
data.columns
```

```python
#посмотрим на количество пропусков в столбцах
data.isna().sum()
```

 - Строк с неизвестным именем и жанром всего 2, поэтому можно их удалить
 - Пропуски в столбцах critic_score, rating, user_score оставим, тк причина их возникновения неизвестна 
 - В столбце year_of_release пропущенные значения заменим на ноль, чтобы преобразовать тип данных в int 

```python
#удалим строки с неизвестным именем
data = data.dropna(subset = ['name'])
```

```python
#посмотрим на отношение пропусков в столбце year_of_release к общему количеству строк
269/data.shape[0]
```

Количество пропусков в столбце year_of_realease меньше 2%, поэтому можно их удалить.

```python
#в year_of_release заменим пропуски на 0, чтобы преобразовать к типу int
data = data.dropna(subset = ['year_of_release'])
data['year_of_release'] = data['year_of_release'].astype(int)
```

```python
#посмотрим уникальные значения в столбце user_score
data['user_score'].unique()
```

tdb(to be determined) = будет определено, значений еще нет, они еще не определены, значит пропуски логичнее всего заменить на NaN.

```python
#в столбце user_score заменим значени tbd на nan
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
data['user_score'] = data['user_score'].astype(float)
```

```python
#поставим заглушку на пропуски в столбцe rating
data['rating'] = data['rating'].fillna('N')
```

```python
#проверим наличие дубликатов по столбцам name, platform, year_of_realese
data.loc[data.duplicated(subset=['name', 'platform', 'year_of_release'], keep=False)]
```

```python
#оставляем первое упоминание
data = data.drop_duplicates(subset=['name', 'platform', 'year_of_release'], keep='first') 
```

```python
#проверим наличие неявных дубликатов в названиях жанров
data['genre'].unique()
```

```python
#посмотрим уникальные значения в столбце rating
data['rating'].unique()
```

 <p>«EC» («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.</p>
 <p>«E» («Everyone») — «Для всех»: Содержание вполне подходит для всех возрастов (ранее с 6 лет). Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первоначально "K-A" ("Kids to Adults")</p>
 <p>«E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»: Проекты с данным рейтингом могут содержать больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года.</p>
 <p>«T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.</p>
 <p>«M» («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику.</p>
 <p>«AO» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое.</p>
 <p>«RP» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.</p>

K-A устаревшее название E, поэтому логично будет их объединить.

```python
#заменим K-A на E
data['rating'] = data['rating'].replace('K-A', 'E')
```

###  Влияние отзывов критиков и пользователей на продажи

```python
#функция для построения зависимостей продаж от отзывов
def depends_on_review(platform_name):
     data = actual_data.query('platform == @platform_name')
    
     fig, axes = plt.subplots(1,2, figsize = (20,8))
     fig.suptitle(f'Зависимость продаж платформы {platform_name} от отзывов', fontsize = 16)
    
     ax1 = plt.subplot(1,2,1)
     data.plot(x='user_score', y='total_sales', grid = True,kind='scatter', ax=ax1)
     plt.xlabel('Отзывы пользователей', size = 14)
     plt.ylabel('Продажи', size = 14)
     plt.title('пользователей', fontsize = 16)

     ax2 = plt.subplot(1,2,2)
     data.plot(x='critic_score', y='total_sales', grid = True,kind='scatter', ax=ax2)
     plt.xlabel('Отзывы критиков', size = 14)
     plt.ylabel('Продажи', size = 14)
     plt.title('критиков', fontsize = 16)
```

```python
#функция для расчета корреляции между отзывами критиков/пользователей и продажами
def corr_sales_score(platform_name):
    data = actual_data.query('platform == @platform_name')
    user_corr = data['user_score'].corr(data['total_sales'])
    critic_corr = data['critic_score'].corr(data['total_sales'])
    display(f'Коэфициент корреляции между отзывами пользователей платформы {platform_name} и продажами равен {user_corr}',f'Коэфициент корреляции между отзывами критиков платформы {platform_name} и продажами  {critic_corr}')
```

```python
corr_sales_score("XOne")
```

- Практически для всех платформ линейная связь между отзывами пользователей и продажами отсутствует. Для платформы 3DS коэффициент линейной связ составляет 0.2, что в разы больше чем у остальных, но эта связь все еще слабая.
- Между отзывами критиков и продажами присутствует линейная связь с коэффициентов корреляции 0.3 - 0.4.
- Коэффициент корреляции для разных платформ отличаются, но не сильно. Продажи напрямую заивисят от отзывов критиков и не зависят от отзывов пользователей.

## Портрет пользователя по регионам

```python
#создадим функцию для расчета топ-5 параметров
def region_param(param, region):
    data_region = actual_data.pivot_table(index = param, values = region, aggfunc = 'sum')
    top_region = data_region.sort_values(by= region, ascending=False).head(5)
    other_index = data_region[region].sum() - top_region[region].sum()
    data_other_index = pd.DataFrame({region : [other_index]}, index=['other'])
    total_data = pd.concat([top_region, data_other_index])
    return total_data
```

### Самые популярные платформы в различных регионах

## Проверка гипотез

```python
#создадим функцию для принятия решения верна ли гипотеза
def deny_or_no(results, alpha):
    if results.pvalue < alpha:
        return ('Отвергаем нулевую гипотезу')
    else:
        return ('Нет оснований отвергнуть нулевую гипотезу')
```

### Нулевая гипотеза средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Альтернативная гипотеза средние рейтинги платформ Xbox One и PC разные(двусторонняя гипотеза). Выборки не заивисимы между собой, поэтому возьмем метод ttest_ind.

```python
xone_rating = actual_data.loc[(actual_data['platform'] == 'XOne') &(actual_data['user_score'] >= 0 )]['user_score']
pc_rating = actual_data.loc[(actual_data['platform'] == 'PC') &(actual_data['user_score'] >= 0 )]['user_score']
platform_results = st.ttest_ind(xone_rating, pc_rating)
display('p-значение:', platform_results.pvalue)
```

```python
pc_rating.mean()
```

```python
xone_rating.mean()
```

```python
deny_or_no(platform_results, 0.05)
```

Вывод: значение pvalue больше значения статистической значимости 0.05, значит нет оснований отвергнуть нулевую гипотезу. Различия рейтингов платформ Xbox One и PC статистически незначимы.

### Нулевая гипотеза средние пользовательские рейтинги жанров Action и Sports  одинаковые

Альтернативная гипотеза средние пользовательские рейтинги жанров Action и Sports разные(двусторонняя гипотеза). Выборки не заивисимы между собой, поэтому возьмем метод ttest_ind.

```python
action_rating = actual_data.loc[(actual_data['genre'] == 'Action')& (actual_data['user_score'] >= 0 )]['user_score']
sports_rating = actual_data.loc[(actual_data['genre'] == 'Sports') & (actual_data['user_score'] >= 0 )]['user_score']
genre_results = st.ttest_ind(action_rating, sports_rating)
display('p-значение:', genre_results.pvalue)
```

```python
action_rating.mean()
```

```python
sports_rating.mean()
```

```python
deny_or_no(genre_results, 0.05)
```

На имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.

**Выводы по разделу**
- данный раздел был посвящен проверке двух гипотез
- в разделе 5.1 была сформулирована нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые; и альтернативная: средние рейтинги платформ Xbox One и PC разные, эта гипотеза является двусторонней, потому что нам важен факт отличия(рейтинг больше или меньше). Выборки не заивисимы между собой, поэтому был выбран метод ttest_ind. Уровень статистической значимости был выбран 0.05. Значение pvalue оказалось больше значения статистической значимости 0.05, значит нет оснований отвергнуть нулевую гипотезу. 
- в разделе 5.2 была сформулирована нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые; и альтернативная: средние пользовательские рейтинги жанров Action и Sports разные. Здесь также двусторонняя гипотеза и независимые выборки, выбрано значение статистической значимости 0.05. Значение pvalue оказалось ниже установленного порога, значит утверждение что средние рейтинги жанров одинаковые является статистически незначимым. Следовательно мы отказываемся от нулевой гипотезы в пользу альтернативной.
